#**2**  Probability and Bayes’ Theorem: Kaggle ML & DS Survey [60pts]

```python
import pandas as pd
import numpy as np

# Load the necessary CSV files
df_multiple_choice = pd.read_csv('multipleChoiceResponses.csv', encoding='latin-1')
df_freeform = pd.read_csv("freeformResponses.csv", encoding ='latin-1')
conversion_rates = pd.read_csv("conversionRates.csv")

```

#**2.a**  [5 pts] What is the probability that a respondent is currently employed as a Programmer given they use C/C++ at work?

```python
# Filter users who use C/C++
cpp_users = df_multiple_choice[df_multiple_choice['WorkToolsFrequencyC'].notnull()]

# Count of users who use C/C++
cpp_users_count = len(cpp_users)

# Count of users among C/C++ users who work as a Programmer
cpp_programmers_count = len(cpp_users[cpp_users['CurrentJobTitleSelect'] == 'Programmer'])

# Calculate the probability
probability_programmer_given_cpp = cpp_programmers_count / cpp_users_count

print("The probability that a respondent is currently employed as a Programmer given they use C/C++ at work is:", probability_programmer_given_cpp)
print("The converted form in percentage:{: .2%}".format(probability_programmer_given_cpp))
```

#**2.b** [5 pts] What is the probability that a respondent is a Data Scientist given they have majored in computer science, mathematics or statistics?

```python
# Filter users who majored in computer science, mathematics, or statistics
relevant_majors = df_multiple_choice[df_multiple_choice['MajorSelect'].isin(['Computer Science', 'Mathematics or statistics'])]

# Count of users who majored in relevant fields
relevant_majors_count = len(relevant_majors)

# Count of users among relevant majors who work as a Data Scientist
data_scientist_count = len(relevant_majors[relevant_majors['CurrentJobTitleSelect'] == 'Data Scientist'])

# Calculate the probability
probability_data_scientist_given_major = data_scientist_count / relevant_majors_count

print("The probability that a respondent is a Data Scientist given they have majored in computer science, mathematics, or statistics is:", probability_data_scientist_given_major)
print("The converted form in percentage: {:.2%}".format(probability_data_scientist_given_major))
```

#**2.c**  What is the probability that a respondent works in the Technology industry given that they earn more than 40,000 USD annually

```python
# Merge with df_multiple_choice to get exchange rates
df_with_rates = pd.merge(df_multiple_choice, conversion_rates, left_on='CompensationCurrency', right_on='originCountry')

# Replace '-' values with NaN
df_with_rates['CompensationAmount'] = df_with_rates['CompensationAmount'].replace('-', np.nan)

# Convert non-USD salaries to USD
df_with_rates['CompensationAmount'] = df_with_rates['CompensationAmount'].astype(str)

# Remove commas and convert to float
df_with_rates['CompensationAmount'] = df_with_rates['CompensationAmount'].str.replace(',', '').astype(float)

# Convert non-USD salaries to USD
df_with_rates['CompensationAmountUSD'] = df_with_rates['CompensationAmount'] * df_with_rates['exchangeRate']

# Filtering those with compensation greater than 40,000 USD
df_filtered = df_with_rates[df_with_rates['CompensationAmountUSD'] > 40000]

# Count of workers in the Technology industry
tech_workers_count = df_filtered[df_filtered['EmployerIndustry'] == 'Technology'].shape[0]

# Total count of workers
total_workers_count = df_filtered.shape[0]

# Probability of being employed in the Technology industry
probability = tech_workers_count / total_workers_count

# Printing the result
print("Probability that a respondent works in the Technology industry given that they earn more than 40,000 USD annually is", probability)
print("The converted form in percentage: {:.2%}".format(probability))
```

#**2.d** What is the joint probability of a respondent being over 30 years old and having a at least a Bachelor’s degree?


```python
# Filter the DataFrame for respondents over 30 years old
over_30_df = df_multiple_choice[df_multiple_choice['Age'] > 30]

# Calculate the total number of respondents over 30
total_over_30 = len(over_30_df)

# Among the respondents over 30, calculate the proportion with at least a Bachelor's degree
bachelors_or_higher_count = len(over_30_df[over_30_df['FormalEducation'].isin(["Bachelor's degree", "Master's degree", "Doctoral degree"])])
joint_probability = bachelors_or_higher_count / total_over_30

# Print the joint probability
print("The joint probability of a respondent being over 30 years old and having at least a Bachelor’s degree is:", joint_probability)
print("The converted form in percentage: {:.2%}".format(joint_probability))
```

#**2.e** What is the probability that a respondent is a Data Scientist who majored in Computer Science, Mathematics or statistics?

```python
# Filter the DataFrame for respondents who are Data Scientists
data_scientist_df = df_multiple_choice[df_multiple_choice['CurrentJobTitleSelect'] == 'Data Scientist']

# Calculate the total number of Data Scientists
total_data_scientists = len(data_scientist_df)

# Among the Data Scientists, calculate the proportion who majored in Computer Science, Mathematics, or Statistics
majored_in_cs_math_stats_count = len(data_scientist_df[data_scientist_df['MajorSelect'].isin(['Computer Science', 'Mathematics or statistics'])])
probability_data_scientist_major_cs_math_stats = majored_in_cs_math_stats_count / total_data_scientists

# Print the probability
print("The probability that a respondent is a Data Scientist who majored in Computer Science, Mathematics, or Statistics is:", probability_data_scientist_major_cs_math_stats)
print("The converted form in percentage: {:.2%}".format(probability_data_scientist_major_cs_math_stats))
```

#**2.f** [5 pts] What is the joint probability that a respondent is from France, earns less than 100,000 USD annually, and uses Cross-Validation Often or Most of the time?

```python
# Filtering participants from France, earning less than 100,000 USD annually, and using Cross-Validation often or most of the time
filtered_df = df_with_rates[(df_with_rates['Country'] == 'France') &
                            (df_with_rates['CompensationAmount'] < 100000) &
                            (df_with_rates['WorkMethodsFrequencyCross-Validation'].isin(['Often', 'Most of the time']))]

# Finding the number of filtered participants
filtered_count = len(filtered_df)

# Finding the total number of participants
total_count = len(df_with_rates)

# Calculating the joint probability
joint_probability = filtered_count / total_count

# Printing the result
print("Joint probability that a respondent is from France, earns less than 100,000 USD annually, and uses Cross-Validation Often or Most of the time:", joint_probability)
print("The converted form in percentage: {:.2%}".format(joint_probability))
```

#**2.g** [10 pts] What is the probability that a respondent uses C/C++ at work given that they are employed as a Programmer? (Hint: Use your findings from Question 2a).

```python
# Prior probability of using C/C++ at work
p_cpp = len(cpp_users) / len(df_multiple_choice)

# Probability of being a programmer given using C/C++
p_programmer_given_cpp = cpp_programmers_count / len(cpp_users)

# Probability of being a programmer
p_programmer = len(df_multiple_choice[df_multiple_choice['CurrentJobTitleSelect'] == 'Programmer']) / len(df_multiple_choice)

# Applying Bayes' theorem
probability_cpp_given_programmer = (p_programmer_given_cpp * p_cpp) / p_programmer

print("The probability that a respondent uses C/C++ at work given that they are employed as a Programmer is:", probability_cpp_given_programmer)
print("The converted form in percentage: {:.2%}".format(probability_cpp_given_programmer))
```

#**2.h** [15 pts] Given the probability of a respondent wearing glasses is 0.15, and the probability of a respondent wearing glasses given they have a PhD is 0.25, find the probability of a respondent having a PhD given that they wear glasses.

```python
# Given probabilities
P_glasses = 0.15  # P(Glass) = 0.15
P_glasses_given_phd = 0.25  # P(Glass | PhD) = 0.25

# Finding the proportion of respondents with a Doctoral degree
total_responses = len(df_multiple_choice)  # Assuming df_multiple_choice is the DataFrame containing the survey data
doctoral_degree_count = df_multiple_choice[df_multiple_choice['FormalEducation'] == "Doctoral degree"].shape[0]
P_phd = doctoral_degree_count / total_responses  # P(PhD)

# Applying Bayes' theorem
P_phd_given_glasses = (P_glasses_given_phd * P_phd) / P_glasses

# Printing the result
print("Probability of having a PhD given that they wear glasses:", P_phd_given_glasses)
print("The converted form in percentage: {:.2%}".format(P_phd_given_glasses))
```
